```python
import requests
import json
import pandas as pd

def pretty(res):
  return json.dumps(res, indent=4, sort_keys=True, ensure_ascii=False)

symbol = input('Digite o código da ação: ')
info = ['dividends', 'fundamentals', 'bpa', 'bpp', 'dfc', 'dre', 'name', 'sector', 'cnpj', 'close', 'stocks']
token = input('Digite o token: ')

response = requests.get('https://api.oplab.com.br/v3/market/stocks/%s?with_financials=%s' % (symbol, ','.join(info)), headers = {'access-token': token, 'accept': 'application/json'})

if response.status_code != 200:
  raise(Exception('Token inválido ou ação não encontrada'))
  
financial = response.json()['financial']
date = '2021-09-30'

l = { k: financial[k] if type(financial[k]) is str else financial[k].get(date, financial[k]) for k in info }

print(pretty(l['fundamentals']))
pl = l['bpp']['2_03']['value']
vpa = pl/l['stocks']['total']
vp = pl
price_on = l['close']['on']
price_pn = l['close']['pn']
stocks_pn = l['stocks']['pn']
stocks_on = l['stocks']['on']
price = (price_on * stocks_on + price_pn * stocks_pn) / (stocks_on + stocks_pn)
p_over_vpa = price/vpa


table = pd.DataFrame([[
  None,
  pl,
  vpa,
  vp,
  p_over_vpa,
  l['fundamentals']['market_cap'],
  l['fundamentals']['ev'],
  None,
  l['fundamentals']['ebit'],
  None,
  l['bpa']['1_01']['value'],
  l['bpp']['2_01']['value'],
  None,
  l['fundamentals']['ev_over_ebit'],

  ]], columns=[
    'Dividend Yield', 
    'PL',
    'VPA',
    'VP',
    'P/VPA',
    'Valor de Mercado',
    'EV',
    'EBITDA',
    'EBIT',
    'Capital de Giro',
    'Ativo Circulante',
    'Passivo Circulante',
    'EV/EBITDA',
    'EV/EBIT',
    # 'P/EBIT',
    # 'P/EBITDA'
    # 'P/Capital de Giro',
    # 'P/Ativo Circulante',
    # 'P/Ativo Total',
    # 'LPA',
    # 'P/SR'
    # 'Dívida Líquida',
    # 'Dívida Líquida/PL',
    # 'Dívida Líquida/EBITDA',
    # 'Dívida Líquida/EBIT',
    # 'PL/Ativo Total',
    # 'Passivo Total/Ativo Total',
    # 'Índice de Liquidez Corrente',
    # 'Margem Bruta',
    # 'Margem EBITDA',
    # 'Margem EBIT',
    # 'Margem Líquida',
    # 'ROE',
    # 'ROA',
    # 'ROIC',
    # 'DRE',
    # 'ROE/EBITDA',
    ])
print(table)
```
